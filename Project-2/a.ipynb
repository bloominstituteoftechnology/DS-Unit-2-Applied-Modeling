```python
features = ['CELLAR', 'BEDROOMS', 'TOTROOMS', 'WINDOWS', 'TOTSQFT_EN']
weights = ['NWEIGHT']
targets = ['btu_log']
```

```python
from xgboost import XGBRegressor
from category_encoders import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

_ohe = OneHotEncoder(   use_cat_names=True, 
						cols=features)
_xgbr = XGBRegressor(   random_state=1,
						n_jobs=-1,
						n_estimators=500,
						max_depth=5,
						learning_rate=.1)

pipeline = Pipeline([
	('OneHotEncoder', _ohe),
	('XGBRegressor', _xgbr)
])

X_train, X_val, y_train, y_val, w_train, w_val = train_test_split(recs[features], recs[targets], recs[weights])

# pipeline.fit(X_train, y_train)
```

```python
print(X_train.shape)
X_train.head()
```

```python
print(w_train.shape)
w_train.head()
```

```python
w_train.values
```

```python
pipeline.fit(   X_train, 
				y_train, 
				XGBRegressor__sample_weight=w_train.values, 
				XGBRegressor__eval_metric='rmse')
```

```python
pipeline.score(X_val, y_val, sample_weight=w_val.values)
```

```python
pipeline.predict(X_val)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
y_pred = pipeline.predict(X_val)
print('R^2 score:', r2_score(y_val, y_pred, sample_weight=w_val.values))
print('RMSE:', numpy.sqrt(mean_squared_error(y_val, y_pred, sample_weight=w_val.values)))
```
